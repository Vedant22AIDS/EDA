```python
df_country=pd.read_excel("Country-Code.xlsx")
```

```python
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df
```

```python
Country_Names=final_df.Country.value_counts().index
```

```python
Country_Names
```

```python
Country_val=final_df.Country.value_counts().values
```

```python
Country_val
```

```python
plt.pie(Country_val[0:3],labels=Country_Names[0:3],autopct='%1.1f%%',explode=[0.2,0.0,0.0])
plt.show()
```

```python
Rating_count=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
Rating_count.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=Rating_count,palette=['white','red','orange','yellow','green'])
```

```python
sns.countplot(x='Rating color',data=Rating_count)
```
